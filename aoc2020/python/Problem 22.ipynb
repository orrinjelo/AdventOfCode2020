````python
with open('../inputs/22.txt') as f:
    lines = f.readlines()

player = None
player1 = []
player2 = []
for line in lines:
    if line.strip() == "Player 1:":
        player = 1
    elif line.strip() == "Player 2:":
        player = 2
    elif line.strip() == '':
        continue
    elif player == 1:
        player1.append(int(line.strip()))
    elif player == 2:
        player2.append(int(line.strip()))

print(player1)
print(player2)
````

````python
def play_round(player1, player2, verbose=False):
    if verbose:
        print("Player 1's deck:", player1)
        print("Player 2's deck:", player2)
        print("Player 1 plays:", player1[0])
        print("Player 2 plays:", player2[0])
    
    p1 = player1[0]
    p2 = player2[0]
    player1.pop(0)
    player2.pop(0)
    
    if p1 > p2:
        if verbose: print("Player 1 wins the round!")
        
        player1.append(p1)
        player1.append(p2)
    else:
        if verbose: print("Player 2 wins the round!")
        
        player2.append(p2)
        player2.append(p1)
````

````python
player1_deck = player1[:]
player2_deck = player2[:]

while len(player1_deck) != 0 and len(player2_deck) != 0:
    play_round(player1_deck, player2_deck)
````

````python
def play_recursive_round(player1, player2, history1=None, history2=None, game=1):
    if history1 is None:
        history1 = []
    if history2 is None:
        history2 = []
    p1 = player1[0]
    p2 = player2[0]
        
    if {x:y for x,y in enumerate(player1)} in history1 or {x:y for x,y in enumerate(player2)} in history2:
        print((p1,p2), "seen before, auto-win for P1!")

        player1.pop(0)
        player2.pop(0)

        player1.append(p1)
        player1.append(p2)
        
        return 1
#         print('p1:', player1)
#         print('p2:', player2)
    else:
        history1.append({x:y for x,y in enumerate(player1)})
        history2.append({x:y for x,y in enumerate(player2)})

        player1.pop(0)
        player2.pop(0)

        if len(player1) >= p1 and len(player2) >= p2:
            player1_copy = player1[:p1]
            player2_copy = player2[:p2]
            play_recursive_game(player1_copy, player2_copy, None, game+1)
            if len(player1_copy) == 0:
                player2.append(p2)
                player2.append(p1)
            else:
                player1.append(p1)
                player1.append(p2)
        else:
            if p1 > p2:
                player1.append(p1)
                player1.append(p2)
            else:
                player2.append(p2)
                player2.append(p1)
    return None

def play_recursive_game(player1, player2, history=None, game=1):
#     print('Game', game)
    if history is None:
        history1 = []
        history2 = []
    r=1
    while len(player1) != 0 and len(player2) != 0:
#         print('Game {}, Round {}'.format(game, r))
#         print('Player 1:', player1)
#         print('Player 2:', player2)
        res = play_recursive_round(player1, player2, history1, history2, game)
        if res:
            break
        r+=1
        
#     print(len(history1), len(history2), r)
    
    if len(player1) == 0:
        player2.reverse()
        p = 0
        for i in range(1,len(player2)+1):
            p += i * player2[i-1]
        return p
    else:
        player1.reverse()
        p = 0
        for i in range(1,len(player1)+1):
            p += i * player1[i-1]
        return p
````
